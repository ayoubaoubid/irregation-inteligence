### 1 Date

Type : object (date)

Description :
Représente la date de l’observation ou de la mesure.
Chaque ligne correspond généralement à un jour de l’année (365 jours).

Utilité :

suivre l’évolution des conditions climatiques

analyser les variations saisonnières

### 2 Temperature

Type : float

Description :
Température de l’air mesurée ce jour-là, généralement en degrés Celsius (°C).

Importance :
La température influence :

l’évaporation de l’eau

la consommation d’eau des plantes

👉 Température élevée → besoin d’irrigation plus important.

### 3 Humidity

Type : float

Description :
Humidité relative de l’air exprimée en pourcentage (%).

Importance :
Elle influence la perte d’eau par évaporation.

Humidity	Effet
faible	plus d’évaporation
élevée	moins d’évaporation
### 4 Rainfall

Type : float

Description :
Quantité de précipitations (pluie) reçue ce jour-là, généralement en millimètres (mm).

Importance :
La pluie contribue naturellement à l’humidité du sol, ce qui peut réduire le besoin d’irrigation.

### 5 ET (Evapotranspiration)

Type : float

Description :
L’évapotranspiration représente la quantité totale d’eau perdue par :

évaporation du sol

transpiration des plantes

Importance :
C’est un indicateur très important pour déterminer les besoins en eau des cultures.

👉 ET élevée → les plantes perdent plus d’eau → irrigation plus nécessaire.

Variables liées à l’humidité du sol
### 6 Soil_Moisture_Smart

Type : float

Description :
Niveau d’humidité du sol mesuré dans un système utilisant l’irrigation intelligente.

Caractéristiques :

contrôlé par capteurs

optimisé par algorithmes ou règles intelligentes

👉 Permet de maintenir un niveau d’eau optimal pour la plante.

### 7 Soil_Moisture_Trad

Type : float

Description :
Niveau d’humidité du sol dans un système utilisant l’irrigation traditionnelle.

Caractéristiques :

irrigation basée sur un calendrier fixe

moins adaptée aux conditions réelles

Variables liées à l’irrigation
### 8 Irrigation_Smart

Type : float

Description :
Quantité d’eau appliquée par le système d’irrigation intelligent.

Fonctionnement :
Le système décide d’irriguer selon :

humidité du sol

météo

evapotranspiration

👉 Objectif : minimiser le gaspillage d’eau.

### 9 Irrigation_Trad

Type : float

Description :
Quantité d’eau appliquée par le système d’irrigation traditionnel.

Caractéristiques :

irrigation régulière

moins sensible aux conditions environnementales

👉 Peut entraîner :

sur-irrigation

gaspillage d’eau.

Variables liées à la santé des cultures
### 10 Crop_Health_Smart

Type : float

Description :
Indicateur de la santé des cultures lorsque l’irrigation intelligente est utilisée.

Interprétation :
Valeur plus élevée → plantes plus saines.

Cet indicateur peut être basé sur :

croissance des plantes

rendement

niveau de stress hydrique.

### 11 Crop_Health_Trad

Type : float

Description :
Indicateur de la santé des cultures avec irrigation traditionnelle.

Objectif :
Permet de comparer les effets des deux méthodes d’irrigation sur la santé des plantes.

### Detection des Valeures Manquantes

### Detection et Traitement des outliers

```python
for column in df.columns[df.dtypes == 'float64']:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = df[column].clip(lower_bound, upper_bound)
```

### Detction et Supprission des doublonts

```python
df = df.drop_duplicates(inplace=True)
```
